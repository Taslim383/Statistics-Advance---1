
1. What is a random variable in probability theory?
→ A variable that takes numerical values based on outcomes of a random experiment.

2. What are the types of random variables?
→ Discrete (countable values) and Continuous (values in an interval).

3. What is the difference between discrete and continuous distributions?
→ Discrete: specific outcomes. Continuous: range of values.

4. What are probability distribution functions (PDF)?
→ PDF gives likelihood of a continuous variable within a range.

5. How do cumulative distribution functions (CDF) differ from PDFs?
→ CDF: probability up to a point. PDF: density at a point (continuous only).

6. What is a discrete uniform distribution?
→ All outcomes equally likely. Example: fair die.

7. What are the key properties of a Bernoulli distribution?
→ Outcomes: 0 and 1. Mean = p, Variance = p(1-p).

8. What is the binomial distribution, and how is it used in probability?
→ Successes in n Bernoulli trials. Used in repeated yes/no experiments.

9. What is the Poisson distribution and where is it applied?
→ Events in a fixed interval. Example: calls per hour.

10. What is a continuous uniform distribution?
→ All values in [a, b] equally likely. PDF = 1/(b-a).

11. What are the characteristics of a normal distribution?
→ Bell-shaped, symmetric, defined by mean and SD.

12. What is the standard normal distribution, and why is it important?
→ Mean = 0, SD = 1. Used for Z-score calculations.

13. What is the Central Limit Theorem (CLT), and why is it critical?
→ Sample means follow normal distribution if sample is large enough.

14. How does the CLT relate to the normal distribution?
→ CLT allows use of normal distribution for sample means.

15. What is the application of Z statistics in hypothesis testing?
→ Measures how far sample mean is from population mean.

16. How do you calculate a Z-score, and what does it represent?
→ Z = (X - μ) / σ. Shows distance from mean in SDs.

17. What are point and interval estimates in statistics?
→ Point: single value. Interval: range (e.g., confidence interval).

18. What is the significance of confidence intervals?
→ Range where parameter is likely to lie, with certain confidence.

19. What is the relationship between a Z-score and confidence interval?
→ Z-score helps compute margin of error in confidence interval.

20. How are Z-scores used to compare different distributions?
→ Standardizes values for comparison across distributions.

21. What are the assumptions for applying the CLT?
→ Independent, random samples, large sample size, finite mean & variance.

22. What is the concept of expected value?
→ Average outcome: E(X) = Σ x·P(x) or ∫ x·f(x) dx.

23. How does a probability distribution relate to expected outcome?
→ Expected value is the average based on distribution's probabilities.
